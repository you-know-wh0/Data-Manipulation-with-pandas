# Setup & Data Loading

```python
import pandas as pd
homelessness = pd.read_csv('/content/homelessness.csv')  # Downloaded the csv file from github ....# Colab stores files in '/content/'
homelessness.head()
```

# Exploring Data

```python
# First few rows
print(homelessness.head())
```

```python
# Check columns and data types
print(homelessness.info())
```

```python
# Summary statistics
print(homelessness.describe())
```

```python
# Number of rows and columns
print(homelessness.shape)
```

```python
# Access values, columns, and index
print(homelessness.values)
print(homelessness.columns)
print(homelessness.index)
```

# Sorting Rows

```python
# Sort by a column(ascending)
homelessness_ind = homelessness.sort_values("individuals")
print(homelessness_ind.head())

# Sort by a column(descending)
homelessness_fam = homelessness.sort_values("family_members", ascending=False)
print(homelessness_fam.head())

# Sort by multiple columns
homelessness_reg_fam = homelessness.sort_values(["region", "family_members"], ascending=[True, False])
print(homelessness_reg_fam.head())
```

# Subsetting Columns

```python
# Single column(returns Series)
individuals = homelessness["individuals"]
print(individuals.head())

# Multiple columns(returns DataFrame)
state_fam = homelessness[["state", "family_members"]]
print(state_fam.head())
```

# Subsetting Rows

```python
# Filter rows where 'individuals' > 5000
homelessness[homelessness['individuals'] > 5000]
```

```python
# Filter rows with a condition
ind_gt_10k = homelessness[homelessness["individuals"] > 10000]
print(ind_gt_10k)

# Filter by categorical value
mountain_reg = homelessness[homelessness["region"] == "Mountain"]
print(mountain_reg)

# Multiple conditions (AND)
fam_lt_1k_pac = homelessness[ (homelessness["family_members"] < 1000) &  (homelessness["region"] == "Pacific")]
print(fam_lt_1k_pac)

# Filter using .isin()
mojave_states = ["California", "Arizona", "Nevada", "Utah"]
mojave_homelessness = homelessness[homelessness["state"].isin(mojave_states)]
print(mojave_homelessness)
```

# Adding New Columns

```python
# Add a new column
homelessness['total_homeless'] = homelessness['individuals'] + homelessness['family_members']
homelessness.head()
```

```python
# Sum two columns
homelessness["total"] = homelessness["individuals"] + homelessness["family_members"]

# Proportion of homeless to state population
homelessness["p_homeless"] = homelessness["total"] / homelessness["state_pop"]

# Homeless individuals per 10k people
homelessness["indiv_per_10k"] = 10000 * homelessness["individuals"] / homelessness["state_pop"]
print(homelessness.head())
```

# Combined Operations

```python
# Add column, filter, sort and subset
homelessness["indiv_per_10k"] = 10000 * homelessness["individuals"] / homelessness["state_pop"]
high_homelessness = homelessness[homelessness["indiv_per_10k"] > 20]
high_homelessness_srt = high_homelessness.sort_values("indiv_per_10k", ascending=False)
result = high_homelessness_srt[["state", "indiv_per_10k"]]
print(result)
```

# Aggregations...Summary Statistics

```python
# Group by 'region' and calculate mean
homelessness.groupby('region')['individuals'].mean()
```

```python
# Mean, median, etc.
print(homelessness["individuals"].mean())
print(homelessness["family_members"].median())

# Custom aggregation (e.g., 30th percentile)
def pct30(column):
    return column.quantile(0.3)

print(homelessness[["individuals", "family_members"]].agg(pct30))

# Multiple aggregations
print(homelessness["total"].agg([min, max, sum, pct30]))
```

# Cumulative Calculations

```python
# Cumulative sum
homelessness["total_cumsum"] = homelessness["total"].cumsum()
print(homelessness[["state", "total_cumsum"]].head())
```
